#### 8. Regression

In statistical modeling, regression analysis is a set of statistical processes for estimating the relationships between a dependent variable and one or more independent variables. (Wiki).

This dataset contains multiple quantitative variables, and the goal of this analysis is to investigate if there is a relation between those variables to each other.  I will use statistical models to estimate a simple relationship between two noisy (meaningless information) sets of observations.

Two main functions in seaborn are used to visualize a linear relationship as determined through regression. These functions are regplot() and lmplot(). lmplot() can be understood as a function that basically creates a linear model plot. lmplot() makes a very simple linear regression plot. Regplot() plots an identical plot except for the size of the plot. The x & y parmameters have to be used for lmplot() while for regplot() the x & y parameters can be entered in various forms.

```python
#add gridlines
sns.set_style('whitegrid') 
#axis total bill & y axis tip
sns.lmplot(x ='total_bill', y ='tip', data = tips) 
```

```python
sns.set_style('whitegrid') 
#axis total bill & y axis tip
reg = sns.regplot(x ='total_bill', y ='tip', data = tips) 
reg.set_title("Regression for total bill vs tip") # title
```

Both are the same except the plots are different sizes

```python
# regression with estimate of central tendency
reg = sns.regplot(x ='size', y ='tip',x_estimator=np.mean, data = tips) 
reg.set_title("Regression for group size vs tip amount") # title
```

```python
# regression with estimate of central tendency
reg = sns.regplot(x ='size', y ='total_bill',x_estimator=np.mean, data = tips) 
reg.set_title("Regression for group size vs total bill") # title
```

As was confirmed above, groups of 5 did not tip enough compared to the amount that would be expected. Before I started this analysis, I would have suspected that as group size increases as does total bill amount and tip amount. A linear model best suits these variables but I would have expected a better fit. Groups of 5 do not tip as would be expected and tables of 5 & 6 did not spend as much pro rata compared to tables sizes of 2, 3 and 4. Simple linear regression is stil the optimal model for these variables.

```python
#regression plot
# 3 variables. Aspect = width of plot, Height = height of plot
sns.lmplot(x="tip", y="total_bill", hue="smoker", aspect=1.7, height=5.5, markers=["x","o"], data=tips)
```

Non- smokers are more predictable and fit a linear regression model much better compared to smokers. While certainly not the 'best fit line', you could have certain confidence in modelling linear regression for non-smokers but not alot for smokers as there is just too much variance in tipping amount to total_bill. The expected tipping amount in the USA is approx. 15%, as per that you would expect tip amount to linearaly increase as total_bill increases to ensure a 15% tip is payed. In the instance of smokers, they regularly tipped >15% while many paid much <15% meaning the average was just >15%. There appeared to be no regularities with how much they tipped.

```python
#regression plot
#4 variables with markers and color changed from default
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", markers=["x","o"], palette="Set2", data=tips)
```

Non-smokers that go to the retaurant at either dinner or lunch do fit a a linear model. Like any real world dataset, it certainly is not a perfect fit to the regression line but non smokers in particular do follow a linear pattern.

```python
#regression plot
#5 variables with markers and color changed from default
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", markers=["x","v"], palette="bright", data=tips)
```

This plot above is meaningful because it breaks key variables up and based on the information we have above, we can make sound assumptions and predictions. It appears that male smokers at dinner are the most unpredictable group for tipping habits. Non-smoker females at lunch on the other hand demonstrated linear relationship with tipping vs the total bill. Females tipping total to bill total appears to be a better linear model compared to males.

```python
#regression plot, 3 variables
sns.lmplot(x="total_bill", y="tip", col="day", aspect=1.3, data=tips,col_wrap=2, height=3) #two plots ont top of another two
```

Thursday and Friday both suit a linear model better than Saturday and Sunday. There were much more customers on the weekend compared to Thursday and Friday which make it more difficult to fit the linear line.
